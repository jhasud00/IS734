```python
def convert_to_numeric(value):
    if value == 'COMPLETE':
        return 3
    elif value == 'PARTIAL':
        return 2
    elif value == 'NONE':
        return 1
    else:
        return 0

def convet_to_numeric(value):
    if value == 'High':
        return 3
    elif value == 'Mediun':
        return 2
    elif value == 'Low':
        return 1
    else:
        return 0
df['access_complexity'] = df['access_complexity'].apply(convet_to_numeric)
df['impact_confidentiality'] = df['impact_confidentiality'].apply(convert_to_numeric)
df['impact_availability'] = df['impact_availability'].apply(convert_to_numeric)
df['impact_integrity'] = df['impact_integrity'].apply(convert_to_numeric)
glucos_col = df['access_complexity']
glucos_col.replace(to_replace = 0, value = glucos_col.median(), inplace=True)
glucose_col = df['impact_confidentiality']
glucose_col.replace(to_replace = 0, value = glucose_col.median(), inplace=True)
BP_col = df['impact_availability']
glucose_col.replace(to_replace = 0, value = BP_col.median(), inplace=True)
st_col = df['impact_integrity']
st_col.replace(to_replace = 0, value = st_col.median(), inplace=True)
```

```python
df['cvss_bin'] = pd.qcut(df['cvss'],
                           q=4,
                           labels=['1', '2', '3', '4'])
```

```python
df.sample(10)
```

```python
input_variables = ['cwe_code','impact_availability','impact_confidentiality','impact_integrity']
x=df[input_variables]
y=df['cvss_bin']
```

```python
testSetSize=0.20
seed=7
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=testSetSize,random_state=seed)
```

```python
LR=LogisticRegression(max_iter=1000)
LR.fit(x_train, y_train)
LR_prdict=LR.predict(x_test)
#DecisionTreeClassifier
DT=DecisionTreeClassifier()
DT.fit(x_train,y_train)
DT_prdict=DT.predict(x_test)
#GaussianNaiveBayes
NB=GaussianNB()
NB.fit(x_train, y_train)
NB_prdict=NB.predict(x_test)
```

```python
#TO-DO 9: Add your code here
print("Accuracy of LR:",metrics.accuracy_score(y_test,LR_prdict))

#Accuracy of DecisionTreeClassifier
print("Accuracy of DT:",metrics.accuracy_score(y_test,DT_prdict))

#NB Accuracy of GaussianNaiveBayes
print("Accuracy of NB:",metrics.accuracy_score(y_test,NB_prdict))
```
